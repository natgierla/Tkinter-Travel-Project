```python
#api wrappers with some given params some optional

#limit will be kept at 20 so it is not being made into a parameter
#api key is the same, it will not be a parameter

def restaurant_call(category, condition, filter_, bias):
    limit='limit=20'
    api_key = "apiKey=c63b3d3792164f3897e37eea32e823c5"
    url = f"https://api.geoapify.com/v2/places?{category}&{condition}&{filter_}&{bias}&{limit}&{api_key}"
    request = requests.get(url)
    rest_json = request.json()
    return rest_json


category = 'categories=catering'
condition = 'conditions=vegetarian'
lat = '40.7685' 
lon = '-73.9822'
filter_ = f'filter=circle:{lon},{lat},1000'
bias = f'bias=proximity:{lon},{lat}'
limit='limit=20'
api_key = "apiKey=c63b3d3792164f3897e37eea32e823c5"


url = f"https://api.geoapify.com/v2/places?{category}&{condition}&{filter_}&{bias}&{limit}&{api_key}"
ex1 = requests.get(url)
json = ex1.json()

rest_list = []

def restaurant_info(rest_json):
    for x in range(5):
        try:
            name = rest_json['features'][x]['properties']['name']
        except:
            name = ''
        try:
            address = rest_json['features'][x]['properties']['formatted']
        except:
            address = ''
        try:
            phonenum = rest_json['features'][x]['properties']['datasource']['raw']['phone']
        except:
            phonenum = ''
        try:
            hours = rest_json['features'][x]['properties']['datasource']['raw']['opening_hours']
        except:
            hours = ''
        try:
            site = rest_json['features'][x]['properties']['datasource']['raw']['website']
        except:
            site = ''
        
        rest_list.append([name, address, phonenum, hours, site])
    
    return rest_list

restaurant_info(json)
```
